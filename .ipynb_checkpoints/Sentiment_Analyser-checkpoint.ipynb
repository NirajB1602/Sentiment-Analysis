```python
# Program for Sentiments Analysis

import csv
punctuation_chars = ["'", '"', ",", ".", "!", ":", ";", '#', '@']


def strip_punctuation(string):
    for char in string:
        if char in punctuation_chars:
            new_str = string.replace(char, "")
            string = new_str
    return string


# lists of words to use
positive_words = []
pos_f = open(r"C:\Users\niraj\Desktop\My Projects\Sentiment Analyser\positive_words.txt")
for lin in pos_f:
    if lin[0] != ';' and lin[0] != '\n':
        positive_words.append(lin.strip())


def get_pos(string):
    str_stripped_p = strip_punctuation(string)
    str_lower_p = str_stripped_p.lower()
    positive_wrd_cnt = 0
    str_lst_p = str_lower_p.split()
    for wrd in str_lst_p:
        if wrd in positive_words:
            positive_wrd_cnt = positive_wrd_cnt + 1
    return positive_wrd_cnt


negative_words = []
pos_f = open(r"C:\Users\niraj\Desktop\My Projects\Sentiment Analyser\_negative_words.txt")
for lin in pos_f:
    if lin[0] != ';' and lin[0] != '\n':
        negative_words.append(lin.strip())


def get_neg(string):
    str_stripped_n = strip_punctuation(string)
    str_lower_n = str_stripped_n.lower()
    negative_wrd_cnt = 0
    str_lst_n = str_lower_n.split()
    for wrd in str_lst_n:
        if wrd in negative_words:
            negative_wrd_cnt = negative_wrd_cnt + 1
    return negative_wrd_cnt


ref_file = open(r"C:\Users\niraj\Desktop\My Projects\Sentiment Analyser\project_twitter_data.csv")
data = ref_file.readlines()

output_file = open(r"C:\Users\niraj\Desktop\My Projects\Sentiment Analyser\resulting_data.csv","w")
output_file.write("Number of Retweets, Number of Replies, Positive Score, Negative Score, Net Score")
output_file.write("\n")
for line in data[1:]:
    sorted_data = line.strip().split(",")
    res_row = ("{},{},{},{},{}".format(sorted_data[1], sorted_data[2], get_pos(sorted_data[0]), get_neg(sorted_data[0]), (get_pos(sorted_data[0]) - get_neg(sorted_data[0]))))
    output_file.write(res_row)
    output_file.write("\n")
output_file.close()
```
